```python
# Project 6: Netflix User Behavior Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
df = pd.read_excel("Dim_User.xlsx")

# Quick check
print("🧾 Columns:", df.columns.tolist())

# Clean column names
df.columns = df.columns.str.strip()

# 📊 1. Genre Popularity
plt.figure(figsize=(10, 6))
genre_counts = df['Genre'].value_counts().head(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="Blues_r")
plt.title("Top Genres Watched")
plt.xlabel("Number of Users")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()

# 🧠 2. Average Weekly Watch Time by Genre
plt.figure(figsize=(10, 6))
genre_time = df.groupby('Genre')['TimeConsumingPerWeek'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=genre_time.values, y=genre_time.index, palette="Greens_r")
plt.title("Average Weekly Watch Time by Genre")
plt.xlabel("Hours per Week")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()

# 👥 3. Device Usage Breakdown
plt.figure(figsize=(8, 6))
device_counts = df['Device'].value_counts()
sns.barplot(x=device_counts.values, y=device_counts.index, palette="Purples")
plt.title("Device Usage for Streaming")
plt.xlabel("Number of Users")
plt.ylabel("Device Type")
plt.tight_layout()
plt.show()

# 📍 4. User Distribution by Country (Top 10)
top_countries = df['Country'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Set2")
plt.title("Top 10 Countries by User Count")
plt.xlabel("Number of Users")
plt.ylabel("Country")
plt.tight_layout()
plt.show()

# 😊 5. Satisfaction Level Distribution
plt.figure(figsize=(8, 5))
satisfaction = df['Satisfaction'].value_counts()
sns.barplot(x=satisfaction.values, y=satisfaction.index, palette="coolwarm")
plt.title("User Satisfaction Levels")
plt.xlabel("Number of Users")
plt.tight_layout()
plt.show()

# 💸 6. Income Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['MontlyIncomeUSD'], bins=10, kde=True, color='orange')
plt.title("Monthly Income Distribution")
plt.xlabel("Income (USD)")
plt.tight_layout()
plt.show()

# 📊 7. Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=8, kde=True, color='teal')
plt.title("User Age Distribution")
plt.xlabel("Age")
plt.tight_layout()
plt.show()

# 🧮 8. Correlation Heatmap (Numeric Fields)
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Age', 'TimeConsumingPerWeek', 'MontlyIncomeUSD', 'FamilyMember']].corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
